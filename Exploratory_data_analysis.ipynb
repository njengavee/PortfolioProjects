```python
# visualize distribution for trip duration
sns.histplot(data['tripduration'])
```

```python
# checking for association between tripduration and gender using scatterplots
sns.pairplot(data[['tripduration', 'gender']])
```

```python
# number of rows that are subscribers
num_subscriber = data['usertype'].loc[data['usertype'] == 'Subscriber'].count()

#number of rows in the dataset
total = len(data)

# Get the percentage of the subscribers
percent_subscriber = round(num_subscriber/total * 100,2)
print('The total number of riders who are annual subscribers is ', percent_subscriber, '%')
```

# Evaluate how trip length varies based on trip start time

```python
data['hour'] = data.starttime.apply(lambda x: x[11:13]).astype('str')
#data

# visualize correlation
sns.scatterplot(x= 'hour', y= 'tripduration', data = data)
```

```python
# visualize correlation
sns.scatterplot(x= 'hour', y= 'tripduration', data = data, hue= 'usertype')
```

# Bike stations where most trips start

```python
# Get the count of trips from each station
new_data = data.groupby(['start station id']).size().reset_index(name= 'counts')
```

```python
#remove duplicate values from the start station id column
temp_data = data.drop_duplicates('start station id')

# left join to merge new_data and temp_data dataframes
newdata2 = pd.merge(new_data, temp_data[['start station id', 'start station name', 'start station latitude', 'start station longitude']], how= 'left', on= ['start station id'])
```

```python
import folium
```

```python
# initialize a map
m = folium.Map(location=[40.691966, -73.981302],tiles= 'OpenstreetMap', zoom_start= 12)
m
```

```python
from matplotlib.patches import Circle
# Add data points as markers on the map
for i in range(0,len(newdata2)):
  folium.Circle(
      location = [newdata2.iloc[i]['start station latitude'], newdata2.iloc[i]['start station longitude']],
      popup = newdata2.iloc[i]['start station name'],
      radius = float(newdata2.iloc[i]['counts'])/2,
      colour = 'blue',
      fill = True, fill_colour = 'blue'
  ).add_to(m)
```
